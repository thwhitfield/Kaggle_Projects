```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import re
import seaborn as sns
import scipy.stats as ss
%matplotlib notebook
pd.set_option('display.max_columns', 500)
sns.set()
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
def get_categorical_columns(df):
    categorical_cols = []
    for col in df.columns:
        if df[col].dtype=='O':
            categorical_cols.append(col)
    return(categorical_cols)
```

```python
def get_categorical_value_counts(df):
    categorical_cols = []
    for col in df.columns:
        if df[col].dtype=='O':
            categorical_cols.append(col)
    df2 = pd.DataFrame()
    for col in categorical_cols:
        my_list = []
        value_counts = list(df[col].value_counts().values)
        value_labels = list(df[col].value_counts().index)
        length = len(value_counts)
        if len(value_counts) > 4:
            my_list = value_counts[:5]
            my_labels = value_labels[:5]
        else:
            my_list = value_counts
            my_labels = value_labels
            for i in range(length,5):
                my_list.append(np.nan)
                my_labels.append(np.nan)
        df2[col] = my_labels
        df2[col+'_value_counts']= my_list
    return(df2)
```

```python
# Another method for determining the correlation between different variables. This would be nice because then I don't have 
# to create "dummy variables" which splits up my data. I can directly compare two columns now. Let's try it.
# It's a little difficult because
def cramers_v(x,y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return (np.sqrt(phi2corr/min((kcorr-1),(rcorr-1))))
```

```python
df_train_cat['SalePriceBins'] = pd.qcut(df_train['SalePrice'],5).astype(str)
```

```python
df = df_train_cat.drop('SalePrice',axis=1)
corr_matrix_cramers = pd.DataFrame(columns=df.columns,index=df.columns)
for col in corr_matrix_cramers.columns:
    for index in corr_matrix_cramers.index:
        corr_matrix_cramers.loc[index,col] = cramers_v(df[col],df[index])

corr_matrix_cramers = corr_matrix_cramers.fillna(0)
fig,ax = plt.subplots(figsize=(12,9))
sns.heatmap(corr_matrix_cramers, vmax=.8, square=True)


# k=10
# cols = corr_matrix_cramers.nlargest(k,'SalePriceBins')['SalePriceBins'].index
# cm = np.corrcoef(df_train_cat[cols].values.T)
# sns.set(font_scale=1.25)
# fig, ax = plt.subplots(figsize=(10,10))
# hm = sns.heatmap(cm,cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':10},
#                  yticklabels=cols.values, xticklabels=cols.values)

# corrmat = df_train.corr()
# fig,ax = plt.subplots(figsize=(12,9))
# sns.heatmap(corrmat, vmax=.8, square=True)
```
